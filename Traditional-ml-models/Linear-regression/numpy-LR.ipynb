```python
import numpy as np
```

```python
area = np.array([[600.0], [1200.0], [1800.0], [2400.0], [3000.0]])
price = np.array([[3200000.0], [7500000.0], [11500000.0], [15500000.0], [21000000.0]])
```

```python
area_mean = area.mean()
area_range = area.max() - area.min()

price_mean = price.mean()
price_range = price.max() - price.min()
```

```python
eps = 1e-8
area_norm = (area - area_mean) / (area_range + eps)
price_norm = (price - price_mean) / (price_range + eps)
```

```python
np.random.seed(42)
w = np.random.randn(1, 1)  
b = np.random.randn(1, 1)
```

```python
lr = 0.05
epochs = 100
n = area_norm.shape[0]
```

```python
for epoch in range(epochs):
    
    # Forward pass: y_pred = wx + b
    pred = area_norm @ w + b   # matrix multiplication
    
    # Compute MSE Loss
    loss = np.mean((pred - price_norm) ** 2)
    
    # Backpropagation (manual gradients)
    # dL/dw = (2/n) * X^T * (pred - y)
    dw = (2/n) * (area_norm.T @ (pred - price_norm))
    
    # dL/db = (2/n) * sum(pred - y)
    db = (2/n) * np.sum(pred - price_norm)
    
    # Update parameters
    w = w - lr * dw
    b = b - lr * db
    
    if (epoch + 1) % 10 == 0:
        print(f"Epoch {epoch+1}, Loss: {loss:.8f}")
```

```python
test_area = np.array([[2000.0]])
test_area_norm = (test_area - area_mean) / (area_range + eps)


pred_norm = test_area_norm @ w + b
```

```python
pred_price = pred_norm * price_range + price_mean
print(f"\nPredicted price: ₹{pred_price.item():,.0f}")
```
