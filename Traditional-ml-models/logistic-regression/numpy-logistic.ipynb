```python
import numpy as np
```

```python
X = np.array([
    [1, 50],
    [2, 55],
    [3, 60],
    [4, 65],
    [5, 70],
    [6, 75]
], dtype=float)

y = np.array([[0], [0], [0], [1], [1], [1]])

```

```python
X.shape
```

```python
X_mean = X.mean(axis=0)
X_range = X.max(axis=0) - X.min(axis=0)

X_norm = (X - X_mean) / X_range
```

```python
print(X_mean)
```

```python
print(X_norm)
```

```python
np.random.seed(42)
w = np.random.randn(2,1)
b = np.random.randn(1)
```

```python
print(w)
print(b)
```

```python
w.shape
```

```python
lr = 0.1
epochs = 7000
n = X.shape[0]
```

```python
n
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
for epoch in range(epochs):

    z = X_norm @ w + b
    
    y_pred = sigmoid(z)

    

    dw = (1/n) * (X_norm.T @ (y_pred - y))
    db = (1/n) * np.sum(y_pred - y)

    w = w - lr * dw
    b = b - lr * db
    loss = -np.mean(y*np.log(y_pred+1e-8) + (1-y)*np.log(1-y_pred+1e-8))

    if epoch % 500 == 0:
        print(f"Epoch {epoch}, Loss: {loss:.4f}")
```

```python
test_hours = np.array([[4.5,68]])
test_norm = (test_hours - X_mean) / X_range

prob = sigmoid(test_norm @ w + b)
prediction = (prob >= 0.5).astype(int)

print("\nProbability of passing:", prob.item())
print("Prediction (0=Fail, 1=Pass):", prediction.item())
```
