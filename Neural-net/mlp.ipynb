```python
import numpy as np
```

```python
X=np.array([[1,2,3,4,5,6],
           [50,55,60,65,70,75]],dtype=float)

Y= np.array([[0,0,0,1,1,1]])

```

```python
n= X.shape[0] #features=2
m= X.shape[1] #samples=6
```

```python
l1=3   #deep layer neuron
l2=1   # Output neuron
W1=np.random.randn(n,l1)*0.01
b1= np.zeros((l1,1))

W2= np.random.randn(l1,l2)*0.01
b2=np.zeros((l2,1))

lr=0.001  #learning rate
epochs= 25

def sigmoid(z):
    return 1/(1+np.exp(-z))

def relu(z):
    return np.maximum(0,z)


def relu_deriv(z):
    return (z>=0).astype(float)
```

```python
for epoch in range(epochs):
    Z1=np.dot(W1.T,X)+b1
    A1=relu(Z1)

    Z2=np.dot(W2.T,A1)+b2
    A2=sigmoid(Z2)

    J = -np.mean(Y*np.log(A2+1e-8) + (1-Y)*np.log(1-A2+1e-8))


    dZ2= A2-Y
    dW2= (1/m)*np.dot(A1,dZ2.T)
    db2=(1/m)*np.sum(dZ2,axis=1,keepdims=True)
    
    dZ1= np.dot(W2,dZ2)*relu_deriv(Z1)
    dW1= (1/m)*np.dot(X,dZ1.T)
    db1=(1/m)*np.sum(dZ1,axis=1,keepdims=True)

    W1 -=lr*dW1
    b1 -=lr*db1

    W2 -=lr*dW2
    b2-=lr*db2

    
    print(epoch,J)


```

```python
x_new=np.array([[4.5],[70]])
```

```python
Z1=np.dot(W1.T,x_new)+b1
A1=relu(Z1)

Z2=np.dot(W2.T,A1)+b2
A2=sigmoid(Z2)

pred =(A2>=0.5).astype(int)

print(A2)

print(pred[0][0])
```
