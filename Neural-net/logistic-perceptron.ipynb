```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))


n,m = X.shape

W = np.zeros((n,1))
b = 0

lr = 0.1
epochs = 1000

for epoch in range(epochs):

    J = 0
    dw = np.zeros((n,1))
    db = 0

    for i in range(m):

        xi = X[:,i:i+1]
        yi = Y[0,i]

        z = W.T @ xi + b
        a = sigmoid(z)

        J += -(yi*np.log(a+1e-8)+(1-yi)*np.log(1-a+1e-8))

        dz = a - yi

        dw += xi * dz
        db += dz

    dw = dw/m
    db = db/m
    J = J/m

    W = W - lr*dw
    b = b - lr*db

    if epoch % 100 == 0:
        print(f"Epoch {epoch} | Loss {J[0][0]:.4f}")
```
