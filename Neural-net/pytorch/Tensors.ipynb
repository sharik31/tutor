##### Mathematical Operations

##### 1. Scalar Operation

```python
X=torch.rand(2,3)
X

Y=torch.rand(2,3)
print(Y)
```

```python
# addition
print(X)
print(Y)
print(X+Y)
# subtraction , # mul , # div , # int div -> //
# mod --> %
# power -> **
```

##### 2. Element Wise Operation

```python
# add , sub, multiply , division , power , mod
# abs-> .abs , # negative -> .neg , #round , ceil, floor 
```

##### 3. Reduction Operations

```python
a= torch.ones(2,3)
a
```

```python
torch.sum(a) # Total sum

# sum along column
print(torch.sum(a, dim=0))
# sum along row
print(torch.sum(a,dim=1))

# mean , # mean along column , # mean along rows
# median 
# max and min
# product
# standard deviation
# variamce

#argmax
#argmin
```

##### 4. matrix Operations

```python
# matmul ,  Transpose , Determinant , Inverse
```

##### 5. special Function

```python
# log , sqrt , sigmoid, softmax-> (a, dim=0),relu
```
